### Introdução a Previsão de Séries Temporais e a Aprendizagem por Reforço [Turma 01B] - 2022/1

```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# Criando a Tabela do Exercício
start = '2020-01'
data = pd.date_range(start,periods=16, freq='M').strftime("%b/%y")
values = [16, 20, 12, 25, 23, 67, 43, 23, 17, 56, 14, 67, 89, 20, 24, 54]
dataset = pd.DataFrame({"Número Série Temporal":values}, index=data)

# Formatando as Colunas 
dataset["Não Centrada"]= dataset.rolling(12, center=True).mean()
dataset["Centrada"]    = dataset['Não Centrada'].rolling(2, center=True).mean()
dataset['Tendência']   = dataset['Não Centrada'] - dataset['Centrada']

# ploting

plt.figure(figsize=(20,8))
plt.title(f"Série Histórica referente a {len(values)} meses", fontsize=20, fontweight='bold')
plt.plot(dataset['Número Série Temporal'], lw=3.0)
plt.ylabel('Valores', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.show()
```
