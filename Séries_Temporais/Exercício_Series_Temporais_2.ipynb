### Introdução a Previsão de Séries Temporais e a Aprendizagem por Reforço [Turma 01B] - 2022/1

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
plt.style.use('fivethirtyeight')
```

```python
# Criando a Tabela do Exercício
start = '2018-01'
data = pd.date_range(start,periods=9, freq='M').strftime("%b/%y")
values = [102,91,95,105,94,101,99,85,101]
dataset = pd.DataFrame({"Número Série Temporal":values}, index=data)

# Formatando as Colunas
pesos = [0.2,0.3,0.5]
dataset["Média Móvel Simples"]= round(dataset.rolling(3).mean(),2).shift(1)
dataset["Média Móvel Ponderada"]= dataset['Número Série Temporal'].rolling(3)\
	.apply(lambda values: np.dot(values, pesos)/np.array(pesos).sum(), raw=True).shift(1)


# Plotando o Gráfico

plt.figure(figsize=(20,8))
plt.title(f"Série Histórica referente a {len(values)} meses", fontsize=20, fontweight='bold')
plt.plot(dataset['Número Série Temporal'], lw=3.0, label = "Série Temporal")
plt.plot(dataset['Média Móvel Simples'], lw=3.0, label = 'Média Móvel Simples')
plt.plot(dataset["Média Móvel Ponderada"], lw=3.0, label= "Média Móvel Ponderada")
plt.ylabel('Valores', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.legend()
plt.show()
```
